```python
job_type = ['job']
```

```python
df_1 = pd.get_dummies(df, columns=job_type, drop_first=True)
```

```python
df_1.head(5)
```

```python
marital_status =['marital']
```

```python
education_type = ['education']
```

```python
df_3 = pd.get_dummies(df_2, columns=education_type, drop_first=True)
```

```python
df_3.head(5)
```

```python
default_type = ['default']
```

```python
df_4 = pd.get_dummies(df_3, columns=default_type, drop_first=True)
```

```python
df_4.head(5)
```

```python
housing_status = ['housing']
```

```python
df_5 = pd.get_dummies(df_4, columns=housing_status, drop_first=True)
```

```python
df_5.head(5)
```

```python
loan_status = ['loan']
```

```python
df_6 = pd.get_dummies(df_5, columns=loan_status, drop_first=True)
```

```python
df_6.head(5)
```

```python
poutcome_type = ['poutcome']
```

```python
df_final = pd.get_dummies(df_6, columns=poutcome_type, drop_first=True)
```

```python
df_final.head(5)
```

```python
df_final.columns
```

```python
#SCaling the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()
```

```python
df_final.dropna()
```

```python
scaler.fit(df_final.drop('y', axis=1))
```

```python
scaled_features = scaler.transform(df_final.drop('y', axis=1))
```

```python
df_fit = pd.DataFrame(scaled_features, columns=df_final.columns[:-1])
```

```python
df_fit
```

```python
#Train_test_split Data 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_fit
```

```python
y= df_final['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train_test_split
```

```python
#Build the KNN model 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model
```

```python
knn_model.fit(X_train,y_train)
```

```python
#Predictions 
```

```python
knn_predictions = knn_model.predict(X_test)
```

```python
knn_predictions
```

```python
X_test
```

```python
len(knn_predictions)
```

```python
test_predictions = pd.DataFrame({'Data_id': X_test.index, 'Real_Credit_Eligibility':y_test,'Credit_eligibility_predictions': knn_predictions})
```

```python
test_predictions.head(20)
```

```python
test_predictions.to_csv('test_predictions_KNN.csv', index=False)
```

```python
#Evaluation
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, knn_predictions))
```

```python
matrix = confusion_matrix(y_test, knn_predictions)
```

```python
sns.heatmap(matrix, cmap='Blues', square=True, annot=True, cbar=False, fmt='d')
```

```python
#Error Rate calculation
```

```python
error_rate=[]

for i in range(1,40):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(X_train, y_train)
    predictions_i = knn_model.predict(X_test)
    error_rate.append(np.mean(predictions_i!=y_test))
```

```python
plt.figure(figsize=(16,7))
plt.plot(range(1,40), error_rate, color='blue', linestyle='dashed', marker='o', markersize=10)
plt.title('Error rate vs. k-value')
plt.xlabel('K-value')
plt.ylabel('Error_rate')
```
