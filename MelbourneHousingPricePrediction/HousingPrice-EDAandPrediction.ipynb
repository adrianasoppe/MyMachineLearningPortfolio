```python
df.insert(0, 'id', range(1, 1 + len(df)))
```

```python
df.head()
```

```python
df.drop(['Address', 'SellerG','Lattitude','Longtitude', 'Date'], axis=1, inplace=True)
df.head()
```

```python
#Encoding Data
```

```python
def Encode(df):
    for column in df.columns[df.columns.isin(['Suburb','Method','CouncilArea','Regionname', 'Type'])]:
        df[column] = df[column].factorize()[0]
    return df

df_en = Encode(df.copy())
```

```python
df_en.head(20)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_en[['id', 'Suburb', 'Rooms', 'Type', 'Method', 'Distance','Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea',
'YearBuilt', 'CouncilArea', 'Regionname', 'Propertycount']]
```

```python
y=df_en['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lm_model = LinearRegression()
```

```python
lm_model
```

```python
lm_model.fit(X_train, y_train)
```

```python
lm_model.intercept_
```

```python
lm_model.coef_
```

```python
coefficients =pd.DataFrame(lm_model.coef_, X.columns, columns=['columns'])
```

```python
coefficients
```

```python
#Predictions
```

```python
predictions = lm_model.predict(X_test)
```

```python
predictions
```

```python
newpredictions = np.round(predictions, 2)
```

```python
newpredictions
```

```python
finalpredictions = pd.DataFrame({'Id': X_test.id, 'Real_price': y_test, 'Predicted_price': newpredictions})
```

```python
finalpredictions
```

```python
finalpredictions.to_csv('Final_predictions_LinearRegression_HousingPrices.csv', index=False)
```

```python
X_test['Predicted_Price'] = newpredictions
```

```python
X_test
```

```python
X_test['Real_price'] = y_test
```

```python
#Evaluation 
```

```python
from sklearn import metrics 
```

```python
print ('MAE:',  metrics.mean_absolute_error(y_test, newpredictions))
```

```python
print ('MSE:',  metrics.mean_squared_error(y_test, newpredictions))
```

```python
print ('RMSE:',  np.sqrt(metrics.mean_absolute_error(y_test, newpredictions)))
```
