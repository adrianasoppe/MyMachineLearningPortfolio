# Titanic Dataset Analysis with Seaborn


## The Data

I will be working with a famous titanic data set for these visualizations. In this notebook I will just focus on the visualization of the data with seaborn:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
plt.figure(figsize=(10,7))
sns.jointplot(data=titanic, x='fare', y='age', color='blue')
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data=titanic, x='fare', color='red', bins=30)
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(data=titanic, x='class', y='age', palette='coolwarm')
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(data=titanic,x='class', y='age', palette='coolwarm')
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=titanic, x='sex', palette='coolwarm')
```

```python
corr = titanic.corr()
```

```python
corr
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(corr, square=True, annot=True, cmap='coolwarm', cbar=False)
plt.title('Titanic corr()')
```

```python
plt.figure(figsize=(25,8))
g = sns.FacetGrid(data=titanic, col='sex')
g = g.map(plt.hist, 'age', bins=20)
```
